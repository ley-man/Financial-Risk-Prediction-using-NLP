```python
# Get names of all sub-directories
import os
folder = "D:/NLP/Data/ACL19_Release"
sub_folders = [name for name in os.listdir(folder) if os.path.isdir(os.path.join(folder, name))]
print(sub_folders)
```

```python
company_details = []
for string in sub_folders:
    temp = string.split('_')
    company_details.append(temp)
    
```

```python
print (company_details)
```

```python
import csv

with open("out.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(company_details)
```
