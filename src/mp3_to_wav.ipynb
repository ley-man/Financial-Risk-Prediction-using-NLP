```python
# Get names of all sub-directories containing mp3 files
folder = r"D:\NLP\data\ACL19_Release"
sub_folders = [name for name in os.listdir(folder) if os.path.isdir(os.path.join(folder, name))]
print(sub_folders)
```

```python
def convert_mp3_to_wav(in_file, out_file):
    subprocess.call(['ffmpeg', '-i', in_file,
                   out_file])
```

```python
# Recursively create directories for wav output files
for file in sub_folders:
    in_path1 = folder
    # Create folder named 'CEO_wav' in the same loc as 'CEO'     
    out_dir_path = in_path1+'/'+file+'/'+'CEO_wav' 
    if not os.path.exists(out_dir_path):
        os.makedirs(out_dir_path)
```
